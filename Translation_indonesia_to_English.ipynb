```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
load_directory = "./translation_model"
tokenizer = AutoTokenizer.from_pretrained(load_directory)
model = AutoModelForSeq2SeqLM.from_pretrained(load_directory)
translator = pipeline("translation", model=model, tokenizer=tokenizer)
indonesian_text = "Bagaimana cuaca hari ini? apa kabarmu"
english_text = translator(indonesian_text)[0]['translation_text']
print(f"Teks Indonesia: {indonesian_text}")
print(f"Teks Inggris (dari model yang dimuat): {english_text}")
```

```python
from transformers import AutoTokenizer
from optimum.onnxruntime import ORTModelForSeq2SeqLM
from pathlib import Path

model_name = "Helsinki-NLP/opus-mt-id-en"
save_directory = Path("./onnx_translation_model")

# Load tokenizer dan konversi model ke ONNX
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = ORTModelForSeq2SeqLM.from_pretrained(model_name, export=True)

# Simpan model dan tokenizer
tokenizer.save_pretrained(save_directory)
model.save_pretrained(save_directory)

print(f"Model ONNX berhasil disimpan di: {save_directory}")
```

```python
from transformers import AutoTokenizer
from optimum.onnxruntime import ORTModelForSeq2SeqLM

# Load tokenizer dan model dari direktori lokal
tokenizer = AutoTokenizer.from_pretrained("./onnx_translation_model", local_files_only=True)
model = ORTModelForSeq2SeqLM.from_pretrained("./onnx_translation_model", local_files_only=True)

# Input teks
text_id = "Bagaimana cuaca hari ini?"

# Tokenisasi input
inputs = tokenizer(text_id, return_tensors="pt")

# Gunakan model ONNX untuk generate output
outputs = model.generate(**inputs)

# Decode hasil terjemahan
translation = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(f"Teks: {text_id}")
print(f"Terjemahan: {translation}")
```
